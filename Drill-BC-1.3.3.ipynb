## Question 1: Choose one variable and plot that variable four different ways.

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
tips = sns.load_dataset("tips")
tips.head(n=10)
```

```python
#Plot 1 showing the varibale 'tip' as a histogram
sns.set(color_codes=True)
sns.set(style="whitegrid", font_scale=1.2)
sns.distplot(tips.tip, axlabel='Tips in US Dollar')
plt.suptitle('Plot 1: Tips as a Histogram')
plt.show()
```

PROS: Here we see the distribution of the variable tip. It also shows the mean value of the tip and the highest value of the tip offered with also showing the variation in the data...it is showing a normal distribution.
CONS: Does not show the spread of individual data points

```python
# Plot 2 showing the varibale 'tip' as a boxplot
box = sns.boxplot(y='tip',data=tips)  
plt.title('Plot 2: Tips as a Boxplot')
box.set( ylabel='Tips in US Dollar')

plt.show()
```

PROS:Shows the Error bars and median very clearly.
CONS: Does not show the spread of individual data points

```python
# Plot 3 showing the varibale 'tip' as a dynamite plot
sns.set(style="darkgrid", font_scale=1.2)
g=sns.factorplot(y="tip", data=tips,
                   size=6, kind="bar", palette="pastel", ci=95)
g.despine(left=True)
g.set_ylabels("Tips in US Dollar")
plt.title('Plot 3: Tips as a dynamite plot')
plt.show()
```

PROS:Shows the mean value clearly and also the eorr bars at CI of 95.
CONS: The bar height from 0 to 2.5 is wasted. It doesnt show the spread of the data.

```python
sct = sns.swarmplot(y='tip', data=tips)
plt.title('Plot 4: Tips as a Scatter plot')
sct.set( ylabel='Tips in US Dollar')
plt.show()
```

PROS:Good for showing the spread of the data down to individual point. good for seeing outliers
CONS: Does not show the mean or median of the data.

## Question 2: Choose two continuous variables, and plot them three different ways.

```python
crashes = sns.load_dataset("car_crashes") # car crashes in 50 states
crashes.head(n=10)
```

```python
#Normalisation to the total number of crashes in each state
crashes['norm_speeding']=crashes.speeding/crashes.total
crashes['norm_alcohol']=crashes.alcohol/crashes.total
crashes.head()
```

```python
# Plot 1: Speeding vs Alcohol as histograms
sns.set(color_codes=True)
sns.set(style="whitegrid")
sns.distplot(crashes.norm_speeding)
sns.distplot(crashes.norm_alcohol, axlabel='No. Crashes with Speeding/Alcohol')
plt.legend(['Speeding', 'Alchol'])
plt.suptitle('Plot 1: Speeding vs Alcohol as histograms')
plt.show()
```

PROS:The distributions  overlap and have different mean values and peak values
CONS: Does not show the spread of individual data points and also no relation can be found here.

```python
# Plot 2: Speeding and Alcohol as boxplots
sns.set(style="whitegrid", font_scale=1.2)
sns.boxplot(y='norm_speeding',data=crashes) 
plt.title('Plot 2: Speeding and Alcohol as boxplots')
box.set( ylabel='No. Crashes with Speeding/Alcohol')
plt.show()
sns.boxplot(y='norm_alcohol',data=crashes, dodge=True) 
plt.show()
```

```python
# Plot 2: Speeding and Alcohol as Scatter plot correlation
g = sns.lmplot(y='norm_speeding', 
               x='norm_alcohol',
               data=crashes, # Data.
               fit_reg=True, # The regression line also includes a 95% confidence envelope.
               scatter_kws={'alpha':0.5})
g.set_ylabels("speeding")
g.set_xlabels(" alcohol")
plt.title('Scatterplot with regression line: Speeding vs Alcohol')
plt.show()
```

```python
# Plot 3: Speeding and Alcohol as Scatter plot with regression line and pearson correlation value

g=sns.jointplot("norm_speeding", "norm_alcohol", data=crashes, kind="reg",
                  color="m")
plt.suptitle('Scatterplot with regression line: Speeding vs Alcohol')
plt.show()
```

No correlation found between speeding and DUI-alcohol from this dataset of car crashes in 50 states

## Question 2: Choose one continuous variable and one categorical variable, and plot them six different ways.

```python
# Loading dataset
diamonds=sns.load_dataset('diamonds')
diamonds.head()
diamonds.carat.max()
diamonds.dtypes
```

```python
# Plot 1: Showing the variables as a boxplot
ax = sns.boxplot(y='price',x='cut',data=diamonds)  
plt.title('Plot 1: Price of diamonds based on cut as a boxplot')
sns.despine(offset=10, trim=True)
ax.set(xlabel='', ylabel='Price in USD')
plt.show()
```

```python
# Plot 2: Showing the variables as a dynamite plot
#sns.set(style="whitegrid")

g = sns.factorplot(x="cut", y="price",  data=diamonds,
                   size=6, kind="bar", palette="pastel", ci=95)
g.despine(left=True)
g.set_ylabels("Price in USD")
plt.title('Plot 2: Price of diamonds based on cut as a dynamite plot')
plt.show()
```

```python
# Plot 3: Showing the variables as a Pointplot

sns.set(style="whitegrid", font_scale=1.2)

g = sns.factorplot(x="cut", y="price", data=diamonds,
                   size=6, kind="point", palette="pastel",ci=95,dodge=True,join=False)
g.despine(left=True)
g.set_ylabels("Price in USD")
plt.title('Plot 3: Price of diamonds based on cut as a Pointplot')
plt.show()
```

```python
# Plot 4: Showing the variables as a Swarmplot/Scatterplot
diamonds=diamonds.sample(n=2000) # This is a huge dataset of 53,940 records so took a random sample to plot of size 2000
swp=sns.swarmplot( x='cut', y='price', data=diamonds)
plt.title('Plot 4: Price of diamonds based on cut as a Swarmplot')
swp.set( ylabel='Price in USD')
plt.show()
```

```python
# Plot 5: Showing the variables as a  Categorical Scatterplot to see if cut is the only factor for determining the price
# For this we need to equal number of each cut variety
diamonds=sns.load_dataset('diamonds')
#limiting the dataframe to only two columns that we need
diamonds=diamonds[['cut', 'price']]
# Taking equal number of  categorical values
diamonds1=diamonds[diamonds['cut']=='Fair'].sample(n=500)
diamonds2=diamonds[diamonds['cut']=='Good'].sample(n=500)
diamonds3=diamonds[diamonds['cut']=='Ideal'].sample(n=500)
diamonds4=diamonds[diamonds['cut']=='Premium'].sample(n=500)
diamonds5=diamonds[diamonds['cut']=='Very Good'].sample(n=500)
diamonds=pd.concat([diamonds1, diamonds2, diamonds3, diamonds4, diamonds5])
diamonds.reset_index(drop=True, inplace=True)
```

```python
diamonds = pd.melt(diamonds, 'cut', var_name='measurement')
swp=sns.swarmplot(x='measurement', y='value', hue='cut', data=diamonds)
plt.title('Plot 5: Price of diamonds based on cut as a Swarmplot')
swp.set( ylabel='Price in USD')
plt.show()
```

From this above plot we see that all five variieties of cutting are spread out along y-axis without any segregation, i.e all colours are mixed up. this indiciates that based on cutting type the price of a diamond cannot be decided.

```python
# Plot 6: Showing the variables as a  Categorical Scatterplot to see if carat is the only factor for determining the price
# For this we need to equal number of each carat variety
diamonds=sns.load_dataset('diamonds')
#limiting the dataframe to only two columns that we need
diamonds=diamonds[['carat', 'price']]
# Taking equal number of  categorical values
diamonds1=diamonds[diamonds['carat'] <= 1.00].sample(n=50)
diamonds2=diamonds[(diamonds['carat'] > 1.00) & (diamonds['carat'] < 2.00)].sample(n=200)
diamonds3=diamonds[(diamonds['carat'] >= 2.00) & (diamonds['carat'] < 3.00)].sample(n=200)
diamonds4=diamonds[(diamonds['carat'] >= 3.00) & (diamonds['carat'] < 4.00)].sample(n=30)
diamonds5=diamonds[(diamonds['carat'] >= 4.00) & (diamonds['carat'] < 5.01)].sample(n=4)

diamonds=pd.concat([diamonds1, diamonds2, diamonds3, diamonds5])
diamonds.reset_index(drop=True, inplace=True)
```

```python
diamonds = pd.melt(diamonds, 'carat', var_name='measurement based on carat')
swp=sns.swarmplot(x='measurement based on carat', y='value', hue='carat', data=diamonds)
plt.title('Plot 6: Price of diamonds based on carat as a Swarmplot')
swp.set( ylabel='Price in USD')
plt.show()
```

From the above plot we can get an idea that carat can be used to determine the price of the diamond. i.e the price is decided based on the carat value of the diamond.
