```python
#importing libraries for analysing data
import pandas as pd
```

```python
# Reading the dataset into a dataframe
df=pd.read_csv("WELLCOME_APCspend2013_forThinkful.csv", encoding='Windows-1251')

# Column name too long...changed to shorter one
df.rename(columns={'COST (Ј) charged to Wellcome (inc VAT when charged)': 'COST (Ј)'}, inplace=True)
df.head()
```

```python
df.info()
```

```python
df['Journal title'].groupby(df['Journal title']).count().sort_values(ascending=False).head(n=20)
```

```python
df['Journal title'].unique()
```

```python
uniq=df['Journal title'].unique()
for x in uniq:
    x=str(x)
    # Replace all journal titles with different cases to only one Title case
    df['Journal title'].replace(to_replace= x, value= x.title(), inplace=True)
    
    
    # Removing all double spaces and making all words to have single spaces between each other
    #and also remove Leading and trailing spaces
    df['Journal title'].replace(to_replace= x, value=' '.join(x.split()), inplace=True)
    
# Replacing certain journal titles like PNAS with abbreviations to expanded form.
uniq=['PNAS', 'P N A S', 'P. N. A. S', 'Pnas']
for x in uniq:
    x=str(x)
    df['Journal title'].replace(to_replace= x, value= 'Proceedings Of The National Academy Of Sciences', inplace=True) 

# REplacing 'Journal' abbrv. to Journal    
df['Journal title'].replace(to_replace= ['J[\s]]*','Jnl[\s]*','Jnl.[\s]*'], value= 'Journal ', inplace=True, regex=True)

#Computing now the question asked... count of common journal names and displaying the most common 5 titles 
#with number of articles published

#pd.set_option('display.max_rows', len(df))
df['Journal title'].groupby(df['Journal title']).count().sort_values(ascending=False).head()
```

```python
# Question: calculate the mean, median, and standard deviation of the open-access cost per article for each journal

# Before doing that the COST column has 'J' as first character in price. 
# Need to remove that for numerical computations
df['COST (Ј)']=df['COST (Ј)'].str.strip((df['COST (Ј)'].loc[0][0:1]))
# or
#df['COST (Ј)']=df['COST (Ј)'].str.strip('Ј') # 'J' is copied from the column name
```

```python
cost_list=[]
for x in df['COST (Ј)']:
    try:
        cost_list.append(float(str(x).strip('$')))
    except ValueError:
        print("Value Errors found", x)
df['COST (Ј)']=pd.DataFrame(cost_list)
df['COST (Ј)']=pd.to_numeric(df['COST (Ј)'])
```

```python
# Mean, median and SD
df.groupby('Journal title', sort=False).mean().rename(columns={'COST (Ј)': 'Mean of COST'}).head(n=10)
```

```python
df.groupby('Journal title', sort=False).median().rename(columns={'COST (Ј)': 'Median of COST'}).head(n=10)
```

```python
df.groupby('Journal title', sort=False).std().rename(columns={'COST (Ј)': 'Standard Deviation of COST'})
```

```python
# Dropping NaN values
df.groupby('Journal title', sort=False).std().rename(columns={'COST (Ј)': 'Standard Deviation of COST'}).dropna()
```

```python
# Mean, median and SD grouped into one output
df.groupby('Journal title', sort=False).agg(['mean', 'median', 'std', 'min', 'max']).rename(columns={'COST (Ј)': 'Mean, Median and Standard dev. of COST'})
```
